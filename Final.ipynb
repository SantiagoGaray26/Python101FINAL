```python
fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(20,10)) # y aqui se pondrian mas columnas para que se vea el comportamiento por semana
index = 0
ax = ax.flatten()

for col, value in ventas_1.items():
    sns.distplot(value, color='g', ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
plt.show()
```
